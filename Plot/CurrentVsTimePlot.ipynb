```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from matplotlib.lines import Line2D
import matplotlib

# 禁用 LaTeX 渲染
matplotlib.rcParams['text.usetex'] = False

# 设置文件夹路径
root = '/Users/jonathanzha/Desktop/Battery-dataset-preprocessing-code-library/NASA/BatteriesChargeCurrent'  # 请替换为实际路径
files = [f for f in os.listdir(root) if f.endswith('.csv')]

# 确保文件按照 Batch1, Batch2, ..., Batch6 的顺序
files.sort()

# 定义颜色和标记
colors = [
    '#80A6E2',
    '#7BDFF2',
    '#FBDD85',
    '#F46F43',
    '#403990',
    '#CF3D3E'
]
markers = ['o', 'v', 'D', 'p', 's', '^']
legends = ['Batch 1', 'Batch 2', 'Batch 3', 'Batch 4', 'Batch 5', 'Batch 6']
line_width = 1.0

# 创建一个2x3的子图
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# 循环遍历所有文件并绘制数据
for i, file in enumerate(files):
    path = os.path.join(root, file)
    data = pd.read_csv(path)
    
    # 确保时间和电压数据是数字类型
    data['Time'] = pd.to_numeric(data['Time'])
    data['Current'] = pd.to_numeric(data['Current'])
    
    ax = axes[i // 3, i % 3]
    
    # 按电池名称和循环寿命分组绘图
    for (battery_name, cycle_life), cycle_data in data.groupby(['Battery Name', 'Cycle Life']):
        label = f'{battery_name} Cycle {cycle_life}'
        ax.plot(cycle_data['Time'], cycle_data['Current'], label=label, alpha=0.7)
    
    # 设置子图标题和标签
    ax.set_title(legends[i])
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Current (A)')
    ax.legend(title='Battery Name & Cycle Life', bbox_to_anchor=(1.05, 1), loc='upper left', frameon=False, fontsize=8)
    ax.grid(True)

# 调整布局
plt.tight_layout()

# 显示图表
plt.show()
```
